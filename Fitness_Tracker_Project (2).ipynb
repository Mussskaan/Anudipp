```python
# Bar chart - Average steps per day

avg_steps_per_day = df.groupby('Data')['TotalSteps'].mean()
plt.figure(figsize=(10,6))
avg_steps_per_day.plot(kind='bar', color='skyblue')
plt.title('Average Total Steps per Day')
plt.ylabel('Steps')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Pie chart - Distribution of very active minutes vs. other minutes

minutes_cols = ['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes']
total_minutes = df[minutes_cols].sum()
plt.figure(figsize=(8,8))
plt.pie(total_minutes, labels=minutes_cols, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of Activity Minutes')
plt.show()
```

```python
# Top 5 days with highest steps

top_steps_days = df[['Data', 'TotalSteps']].sort_values(by='TotalSteps', ascending=False).head()
print("\nTop 5 days with highest total steps:")
print(top_steps_days)
```

```python
plt.figure(figsize=((10,6)))
plt.hist(df.DayOfWeek,color='gray',bins=7,width =.6,align='mid')

plt.xlabel('Day of Week')
plt.ylabel('freq')
plt.title('freq day of week')

plt.grid()
plt.show()
```

```python
mask = df.drop(['DayOfWeek','Data'],axis=1)
```

```python
mask.corr()
```

```python
sns.heatmap(mask.corr())
```

```python
plt.scatter(mask.TotalSteps,mask.Calories,c=mask.Calories )

median_steps = 7405
median_calories = 2134

plt.axhline(median_calories,color = 'b',label = 'median of calories')
plt.axvline(median_steps,color = 'r',label = 'median of steps')

plt.xlabel('Steps')
plt.ylabel('Calories')
plt.title('Calories & TotalSteps')

plt.legend()
plt.show()
```

```python
plt.scatter(mask.TotalHours,mask.Calories,c=mask.Calories )

median_Hours = 24
median_calories = 2134

plt.axhline(median_calories,color = 'b',label = 'median of calories')
plt.axvline(median_Hours,color = 'r',label = 'median of Hours')

plt.xlabel('Hours')
plt.ylabel('Calories')
plt.title('Calories & Total Hours')

plt.legend()
plt.show()
```

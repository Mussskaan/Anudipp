```python
import matplotlib.pyplot as plt
manufacturers=["samsung","apple","oppo","vivo","others"]
market_share=[30,50,56,13,50]
plt.pie(market_share,labels=manufacturers)
plt.title("smartphone market share")
plt.show()
```

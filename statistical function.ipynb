```python
#exponential and logarithms
```

```python
import numpy as np
```

```python
arr=np.array([1,2,3,4,5])
print("exponentiation(e^x):",np.exp(arr))
print("natural logarithm(ln):",np.log(arr))
```

```python
matrix_a=np.array([[1,2],[3,4]])
matrix_b=np.array([[5,6],[7,8]])
print("matrix multiplication:",np.dot(matrix_a,matrix_b))
print("original matrix:",matrix_a)
print("after transposition:",np.transpose(matrix_a))
```

```python
#rounding and truncation
```

```python
data=np.array([3.14759,2.71828,1.6158])
print("rounded to 3 decimals:",np.round(data,3))
print("truncated:",np.trunc(data))
```

```python
#real time example                  
```

```python
monthly_contribution=1000
annual_interest_rate=6.0
years=20
monthly_interest_rate=(annual_interest_rate/12)/100
n=years*12
future_value=monthly_contribution*((1+monthly_interest_rate)**n-1)/monthly_interest_rate
print("future value of investment:",round(future_value,2))
```

```python
#numpy logical operations
```

```python
arr1=np.array([1,2,3])
arr2=np.array([1,5,3])
print("equal:",arr1==arr2)
print("greater than:",arr1>arr2)
print("array-wise equal(arr1 vs arr2):",np.array_equal(arr1,arr2))


```

```python
temperature=np.array([52,84,24,56,32])
rainfall=np.array([5,0,10,0,8,15])
hot_days=temperature>30
rainy_days=rainfall>0
print("hot days:",hot_days)
print("rainy day:",rainy_days)
```

```python
genres=["mystery","romance","science fiction","fantasy","thriller"]
books_sold=[120,90,80,110,70]
plt.bar(genres,books_sold)
plt.xlabel("genre")
plt.ylabel("number of books sold")
plt.title("books sold by genre over a year")
plt.show()
```

```python
age_groups=["0-10","11-20","21-30","31-40","41-50","51-60","61-70","71+"]
population=[15000,22000,30000,28000,2500,18000,12000,8000]
plt.barh(age_groups,population)
plt.ylabel("age group")
plt.ylabel("population distribution by age group")
plt.show()
```
